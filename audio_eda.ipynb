```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(3)
```

```python
df_0 = pd.read_csv('./fluent_speech/data/speaker_demographics.csv', index_col=0)
df_0 = df_0[['gender']]
df_0
```

```python
df_0.gender.value_counts()
```

```python
df_1 = pd.read_csv('./fluent_speech/data/train_data.csv', index_col=2)
df_1 = df_1[['path']]
df_1
```

```python
df = df_1.copy()
df = df.reset_index()
df['gender'] = df.speakerId.map(lambda x: str(df_0.loc[x][0]))
df
```

```python
df.gender.value_counts()
```

```python
delta = df.gender.value_counts()[0] - df.gender.value_counts()[1]
index = df[df.gender == 'male'].index
drop_indices = np.random.choice(index, delta, replace=False)
df_bal = df.drop(drop_indices)
df_bal
```

```python
df_bal.gender.value_counts()
```

```python
df.to_csv('./fluent_speech/index.csv', index=False)
```
